```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ai_job_dataset1.csv") 
df.head()
```

```python
df.shape                   # Rows & columns
df.columns                 # Columns names
df.info()                  # Data types + non-null counts
df.describe()              # Summary stats for numeric columns
```

```python
df.isnull().sum().sort_values(ascending=False)             # Idendify missing values
```

```python
df['salary_usd'] = df['salary_usd'].replace('[\$]', '', regex=True).astype(float)
```

```python
df['company_location'].unique()
df['job_title'].value_counts().head(10)
```

```python
df['compant_location'] = df['company_location'].str.strip().str.title()
df['job_title'] = df['job_title'].str.lower().str.strip()
```

```python
df['experience_level'] = df['experience_level'].str.lower().str.strip()
df['employment_type'] = df['employment_type'].str.lower().str.strip()
```

```python
experience_map = {
                'entry-level' : 0,
                'mid_level' : 1,
                'senior-level' : 2,
                'executive' : 3,
}
df['experience_numeric'] = df['experience_level'].map(experience_map)
```

```python
df.to_csv("ai_jobs_2025_cleaned.csv", index=False)
```
